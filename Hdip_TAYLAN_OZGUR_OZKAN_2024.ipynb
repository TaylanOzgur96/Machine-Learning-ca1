```python
# Filtering outliers with the IQR method

# Identify outlier boundaries for each continuous variable and remove them from the data set.
for column in continuous_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filtering outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# New dimensions of the dataset after outlier cleaning
df.shape
```

## Cluster Analysis

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# We apply Label Encoding to numericalize categorical variables
label_encoder = LabelEncoder()
df_encoded = df.copy()  # We create a copy to preserve the original data

# Encoding categorical columns
categorical_columns = ["Country", "Region", "Crop_Type", "Adaptation_Strategies"]
for column in categorical_columns:
    df_encoded[column] = label_encoder.fit_transform(df[column])

# Creating K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded["Cluster"] = kmeans.fit_predict(df_encoded)

# Show the first few rows of a numeric dataset
df_encoded.head()
```

### Determining the Number of Clusters

```python
import numpy as np

# Determining the optimal number of clusters with the Elbow Method
wcss = []  # List of within-cluster total squared deviations
K = range(1, 11)  # We test clusters 1 to 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_encoded)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow graph
plt.figure(figsize=(10, 6))
plt.plot(K, wcss, marker="o")
plt.title("Elbow Method for Optimal Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.show()
```

```python
df_encoded["Cluster"].value_counts()
```

### Running the Algorithm and Visualizing the Results

```python
# We will perform dimensionality reduction with PCA to visualize the clustering results with a 2D scatter plot.
from sklearn.decomposition import PCA

# 2D dimension reduction with PCA (we reduce it to 2 components)
pca = PCA(n_components=2)
pca_components = pca.fit_transform(df_encoded.drop(columns="Cluster"))

# We can differentiate the clusters by adding the PCA results to a DataFrame.
df_encoded["PCA1"] = pca_components[:, 0]
df_encoded["PCA2"] = pca_components[:, 1]

# Visualizing clusters with a 2D scatter plot
plt.figure(figsize=(10, 8))
for cluster in df_encoded["Cluster"].unique():
    cluster_data = df_encoded[df_encoded["Cluster"] == cluster]
    plt.scatter(cluster_data["PCA1"], cluster_data["PCA2"], label=f"Cluster {cluster}", alpha=0.6)

plt.title("K-Means Clustering Results in 2D (PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Reduction to 3 components with PCA
pca_3d = PCA(n_components=3)
df_encoded_pca_3d = pca_3d.fit_transform(df_encoded.drop("Cluster", axis=1))

# Let's add the data reduced with PCA to the DataFrame
df_encoded["PCA3"] = df_encoded_pca_3d[:, 2]
```

```python
# 2D Graph: PCA2 vs PCA3
plt.figure(figsize=(10, 8))
scatter = plt.scatter(df_encoded["PCA2"], df_encoded["PCA3"], c=df_encoded["Cluster"], cmap="viridis", alpha=0.6)
plt.title("K-Means Clustering Results in 2D (PCA2 vs PCA3)")
plt.xlabel("Principal Component 2")
plt.ylabel("Principal Component 3")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

```python
# 3D Graph: PCA1, PCA2 and PCA3
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df_encoded["PCA1"], df_encoded["PCA2"], df_encoded["PCA3"], c=df_encoded["Cluster"], cmap="viridis", alpha=0.6)
ax.set_title("K-Means Clustering Results in 3D (PCA Reduced)")
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

## Time Series Analysis and ARIMA Model

```python
# Stationarity Check with ADF Test
from statsmodels.tsa.stattools import adfuller

# Let's take the average temperature data on a "Year" basis and make it suitable for time series.
temperature_ts = df.groupby("Year")["Average_Temperature_C"].mean()
```

```python
# Prepare for trend and seasonality analysis by visualizing time series
plt.figure(figsize=(12, 6))
temperature_ts.plot(title="Average Temperature Over Years")
plt.xlabel("Year")
plt.ylabel("Average Temperature (C)")
plt.show()
```

```python
adf_test = adfuller(temperature_ts)
print("ADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plotting ACF and PACF graphs with 8 lags
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_acf(temperature_ts, lags=8, ax=plt.gca())
plt.title("ACF - Autocorrelation Function")

plt.subplot(1, 2, 2)
plot_pacf(temperature_ts, lags=8, ax=plt.gca())
plt.title("PACF - Partial Autocorrelation Function")

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Defining and training the ARIMA model
model = ARIMA(temperature_ts, order=(1, 0, 1))
arima_result = model.fit()
```

```python
arima_result.summary()
```

```python
# Making predictions for the next 5 years
forecast = arima_result.forecast(steps=5)

# Labeling forecast results with actual years
last_year = temperature_ts.index[-1]  # Last year in current time series
forecast.index = range(last_year + 1, last_year + 6)  # Adjusting forecasts to subsequent years
print(forecast)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Divide data into train and test sets for evaluation
train_data = temperature_ts[:-5]  # Keep the last 5 points for testing
test_data = temperature_ts[-5:]

# Fit the ARIMA model on the training data
model = ARIMA(train_data, order=(1, 0, 1))
arima_fit = model.fit()

# Forecast for the test period (last 5 years)
forecast = arima_fit.forecast(steps=5)

# Calculate performance metrics
mae = mean_absolute_error(test_data, forecast)
mse = mean_squared_error(test_data, forecast)
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Visualization of actual and forecasted values
plt.figure(figsize=(10, 6))

# Plot actual data
plt.plot(temperature_ts, label="Actual Values", color="blue")

# Plot forecasted values (extend line to show forecast years clearly)
plt.plot(test_data.index, forecast, label="ARIMA Forecast", linestyle="--", color="orange")

# Add titles and labels
plt.title("ARIMA Model Forecast vs. Actual Values")
plt.xlabel("Year")
plt.ylabel("Average Temperature (C)")
plt.legend()
plt.grid(True)
plt.show()
```
